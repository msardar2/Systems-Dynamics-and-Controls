```matlab
%This is Example 4 (PD Controller Design) from the Root Locus Playlist. CAN Education on YouTube.
t = 0:0.001:3
```

```matlab
s = tf('s')
```

```matlab
Gp1 = 1/(s*(s+2))
```

```matlab
Gc1 = 1
```

```matlab
G1 = Gp1*Gc1
```

```matlab
H1 = feedback(Gc1*Gp1,1)
```

```matlab

Gc2 = (2*s+8)
```

```matlab
G2 = Gp1*Gc2
```

```matlab
H2 = feedback(Gc2*Gp1,1)
```

```matlab
stepinfo(H2) %Max Overshoot is good but Ts is greater than the required. Tunning required
```

```matlab

Gc3 = (s+4)*30.789
```

```matlab
G3 = Gp1*Gc3
```

```matlab
H3 = feedback(Gc3*Gp1,1)
```

```matlab
%sisotool(G3)
stepinfo(H3)
```

```matlab

%Interpret the graphs according to table in Study Note 2
step(H1,t)
title("No Controller Step Response")
```

```matlab
step(H2,t)
title("PI Controller Step Response")
```

```matlab
step(H3,t)
title("Tuned Step Response")
```
