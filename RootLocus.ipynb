```matlab
% steady state error
close all
clc
g1=tf(3,[1 3]);
g2=tf(3,[1 3 0]);
g3=tf([1 1],[1 3 0 0]);
h1=feedback(g1,1);
h2=feedback(g2,1);
h3=feedback(g3,1);
h=h3 %specify system type
```

```matlab
tmax=50
```

```matlab
t=0:0.01:tmax;
us=ones(1,length(t));
ur=t;
up=0.5*(t.^2);;
u=up %specify input
```

```matlab
y=lsim(h,u,t);
e=zeros(1,length(t));
for i=1:length(t);
e(i)=u(i)-y(i);
end
figure (1)
plot(t,y)
hold on
plot(t,u,'r')
title('response')
grid on
```

```matlab
figure (2)
plot(t,e)
title('error')
grid on
```
